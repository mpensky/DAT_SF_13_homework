**Data Set Information:**

The "spam" concept is diverse: advertisements for products/web sites, make money fast schemes, chain letters, pornography...

**Attribute Information:**

The last column of 'spambase.data' denotes whether the e-mail was considered spam (1) or not (0), i.e. unsolicited commercial e-mail. Most of the attributes indicate whether a particular word or character was frequently occuring in the e-mail. The run-length attributes (55-57) measure the length of sequences of consecutive capital letters.

###Definitions of the attributes:

**48 continuous real [0,100] attributes of type word_freq_WORD**

= percentage of words in the e-mail that match WORD, i.e. 100 * (number of times the WORD appears in the e-mail) / total number of words in e-mail. A "word" in this case is any string of alphanumeric characters bounded by non-alphanumeric characters or end-of-string.

**6 continuous real [0,100] attributes of type char_freq_CHAR**

= percentage of characters in the e-mail that match CHAR, i.e. 100 * (number of CHAR occurences) / total characters in e-mail

**1 continuous real [1,...] attribute of type capital_run_length_average**

= average length of uninterrupted sequences of capital letters

**1 continuous integer [1,...] attribute of type capital_run_length_longest**

= length of longest uninterrupted sequence of capital letters

**1 continuous integer [1,...] attribute of type capital_run_length_total**

= sum of length of uninterrupted sequences of capital letters = total number of capital letters in the e-mail

**1 nominal {0,1} class attribute of type spam**

= denotes whether the e-mail was considered spam (1) or not (0), i.e. unsolicited commercial e-mail.

###Exercise 1: Explore the dataset and display some visualizations showing how the variables relate to each other

```python
# Correlation heat map
sns.set(style="darkgrid")

data4corr = spam_data.ix[:,0:-1]

f, ax = plt.subplots(figsize=(20, 20))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.corrplot(data4corr, annot=False, sig_stars=False,
             diag_names=False, cmap=cmap, ax=ax)
f.tight_layout()

# Interesting behavior in the data where there are higher correlations across some word frequencies, like some numbers (415
# vs 857) and then across some words (free vs. 000 and address). Seems like top section may relate more to spam and middle
# may relate more to non-spam
```

###Exercise 2: Build a simple logistic regression and visualize it
use the variable "capital_run_length_longest" to predict "is_spam" How accurate is this single feature?

```python
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

main_feature = spam_data[['capital_run_length_longest']]
target = spam_data.is_spam.values

features_train, features_test, target_train, target_test = train_test_split(main_feature, target, test_size=0.30, random_state=0)
lr = LogisticRegression(C=1)
lr.fit(features_train, target_train)
target_predicted = lr.predict(features_test)

#evaluate accuracy
print("\n\nLogistic regression of Spam dataset on just capital_run_length_longest\n\n")

print(classification_report(target_test, target_predicted,
                           target_names=['Not spam', 'Spam']))

# The results show that just using the longest capital running set is fairly predictive. It has an F1 score of 0.7. It is
# best at recalling not spam (it recognized 94% of truly not spam as being not spam) and worst at recalling spam 
# (it said that only 40% of truly spam is spam). Precision is better and more even (70% of what was recognized as not 
# spam was actually not spam, and 81% of what was recognized as spam is spam). 
```

###Exercise 3: Use train-test spit to split your data at a 30% mark and run another logistic regression using all variables
use random state = 12 so that we can compare results

```python
allfeatures = spam_data.ix[:,0:-1]
target = spam_data.is_spam.values

features_train, features_test, target_train, target_test = train_test_split(allfeatures, target, test_size=0.30, random_state=12)
lr = LogisticRegression(C=1)
lr.fit(features_train, target_train)
target_predicted = lr.predict(features_test)

#evaluate accuracy
print("\n\nLogistic regression of Spam dataset on just capital_run_length_longest\n\n")

print(classification_report(target_test, target_predicted,
                           target_names=['Not spam', 'Spam']))

# Results when using all predictors improves a lot! Precision/Recall are all in the mid-90%'s and f1 score is also much higher
```

###Exercise 4: Apply cross validation to see how the model fares across different splits of your data
Use cross validation to score your model

```python
from sklearn.cross_validation import cross_val_score

# LR at various splits of the data
lr = LogisticRegression(C=1)
lr.fit(allfeatures, target)
lrscores = cross_val_score(lr, allfeatures.values, target, cv=5) #Here, using 5 folds
print lrscores
print 'Mean score is ' + '{:.2f}%'.format(np.mean(lrscores)*100) + ' with standard deviation of ' + '{:.2f}%'.format(np.std(lrscores)*100)
```

###Exercise 5: Compare Performance of Logistic Regression to KNN with 3 neighbors
Which model is more accurate?

```python
# Below we'll run cross-validation on KNN with 3 neighbors to compare with the results from Exercise 4

from sklearn import neighbors

# Get KNN classifier at various cross validations
clf_xv = neighbors.KNeighborsClassifier(3, weights='uniform') #Initialize KNN
clf_xv.fit(allfeatures, target) 
skscores = cross_val_score(clf_xv, allfeatures.values, target, cv=5) #Here, using 5 folds
print skscores
print 'Mean score is ' + '{:.2f}%'.format(np.mean(skscores)*100) + ' with standard deviation of ' + '{:.2f}%'.format(np.std(skscores)*100)

# KNN is much less accurate than logistic regressions here (mean score is only 78% vs. 91% with Logistic regression)
```

###Exercise 6: Evaluate Feature Importance
Which features are the most influential in this model?

```python
# We will use a decision tree classifier to tell us which featuers seem to be most important. 

# First, figure out what max depth to use wih the decision tree.

from sklearn.tree import DecisionTreeClassifier
meanscores=[]
for md in range(1,15):
    treeclf = DecisionTreeClassifier(max_depth=md, random_state=1)
    scores = cross_val_score(treeclf, allfeatures.values, target, cv=5)
    meanscores.append(np.mean(scores))

plt.plot(range(1,15),meanscores)
plt.xlabel('Max Depth')
plt.ylabel('Scores')

# Looks like max depth of 6 or 10 seems to provide best results. Will stick with 6 to avoid overfitting.
```

```python
treeclf = DecisionTreeClassifier(max_depth=5, random_state=1) 
treeclf.fit(allfeatures.values, target)

# compute the feature importance
impts = pd.DataFrame({'feature':allfeatures.columns, 'importance':treeclf.feature_importances_}).sort('importance',ascending=False).reset_index().ix[:20]
impts

# It looks like the first 14 variables in the list below actually matter. The rest are superflous. Also, not too surprisingly,
# the frequency of $, remove and ! matter a lot to decide on whether this is spam or not. Gini likely split out $ as the
# first feature because it explains so much of the variation between 1's and 0's in the is_spam target array. 
```

###Exercise 7: Plot the ROC Curve for the logistic regression you chose¶

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import auc

mostimportantfeatures = spam_data[impts.feature[:15].values]

features_train, features_test, target_train, target_test = train_test_split(mostimportantfeatures, target, test_size=0.30, random_state=12)
lr = LogisticRegression(C=1)
lr.fit(features_train, target_train)
predicted_proba = lr.predict_proba(features_test)

def plot_roc_curve(target_test, target_predicted_proba):
    fpr, tpr, thresholds = roc_curve(target_test, target_predicted_proba[:, 1])   
    roc_auc = auc(fpr, tpr)
    # Plot ROC curve
    plt.plot(fpr, tpr, label= 'ROC Area = %0.3f' % roc_auc)
    plt.plot([0, 1], [0, 1], 'k--')  # random predictions curve
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.xlabel('False Positive Rate or (1 - Specifity)')
    plt.ylabel('True Positive Rate or (Sensitivity)')
    plt.title('ROC')
    plt.legend(loc="lower right")
    
plot_roc_curve(target_test, predicted_proba)
```

###Exercise 8: Demonstrate how the accurary of your predictions changes when you set your threshold to levels other than 50%

```python
predicted_proba = lr.predict_proba(features_test)
predicted_proba_spam = predicted_proba[:,1]

acc = dict()
for thresh in np.linspace(0., 1, 11):
    pred = 1*(predicted_proba_spam>thresh)
    acc[thresh] = 1 - np.mean(abs(pred - target_test))
```

```python
pd.Series(acc).plot()
plt.xlabel('Threshold')
plt.ylabel('Accuracy of identifying spam in test set')

# The graph below shows that having a threshold between 0.4 and 0.6 likely leads to the highest accuracy. 
# 0.4 leads to the highest accuracy, but 0.5 and 0.6 are about the same. 

print pd.Series(acc)
```

###Exercise 9: Discuss the pro's/con's of moving the threshold away from 50%, why is/isn't this a good idea?

```python
# Pros: If you want to have the spam filter be more sensitive, you can reduce the threshold. This will catch more spam
# and reduce the amount of "junk" mail received. You could instead increase the threshold and catch less spam but have 
# fewer false positives where normal email is characterized as spam. This will ensure that people receive the email that
# they need. 

# Cons: Reducing the threshold will sometimes characterize normal email as spam and users will not see it. In the other 
# direction, users will get their email, but will have a lot more spam in their box. 
```

```python
# Given this application, it's probably best to err on the side of a higher threshold. People can delete any missed spam
# emails, but will be unhappy if necessary emails are characterized as spam and not seen. 
```
