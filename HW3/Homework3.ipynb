```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
```

```python
# 1. Use the file bank.csv to explore the dataset. Observe the features: Are they numbers? Are they strings? 
# Are they binary? Are they continuous?

bankdata = pd.read_csv('bank.csv',sep=';')
bankdata.head(10)
```

```python
bankdata.isnull().sum()
```

```python
bankdata.info()
# The table below shows the features and their types. The features are either ints or objects (strings). 
# There are no floats in the data.  Most of the features are discontinuous/unordered, except age, balance, duration 
# and pdays.
```

```python
#2. Learn about label encoders at the following link and use what you learn to transform the features to numerical 
# features.
# http://scikitlearn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html

from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder()

# OneHotEncoder only takes integer matrices, which is a big frustrating. First, have to convert our categorial features to
# integers

def text2int(DF, column): #Converts columns from text to strings
    text = DF[column].drop_duplicates().values
    toints = range(0,len(text))
    mydict = dict(zip(text,toints))
    DF[column]=DF[column].replace(mydict)
    return DF

categoricalvals = ['job','marital','education','default','housing','loan','contact','day','month','poutcome','y']
for col in categoricalvals:
    bankdata = text2int(bankdata, col)
    
bankdata.head()
```

```python
a = enc.fit_transform(bankdata[['job','marital','education','default','housing','loan','contact','day','month','poutcome']])  
dummies = pd.DataFrame(a.toarray()) 

#Is this really better than using pd.get_dummies? It seems simpler and at least the data is labeled there...
#Also, you can remove one column there so you avoid multicollinearity problems.
```

```python
X = pd.concat([bankdata[['age','balance','duration','campaign','pdays','previous']],dummies],axis=1)
y = bankdata.y.values
```

```python
#3. Build a decision tree model to predict whether a prospect will buy the product.
# Using max depth of 3 here... 
from sklearn.tree import DecisionTreeClassifier
treeclf = DecisionTreeClassifier(max_depth=3, random_state=1) #random_state = seed value
treeclf.fit(X, y)

# compute the feature importance
pd.DataFrame({'feature':X.columns, 'importance':treeclf.feature_importances_}).sort('importance',ascending=False).reset_index().ix[:10]
```

```python
# 4. Evaluate the accuracy of your decision tree model using cross validation.

# Going to try out various max depths to see which one is best
from sklearn.cross_validation import cross_val_score
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

meanscores=[]
for md in range(1,15):
    treeclf = DecisionTreeClassifier(max_depth=md, random_state=1)
    scores = cross_val_score(treeclf, X, y, cv=5)
    meanscores.append(np.mean(scores))

plt.plot(range(1,15),meanscores)
plt.xlabel('Max Depth')
plt.ylabel('Scores')

# Looks like the model is pretty accurate (~90%) and has highest accuracy at a max depth of 3 or 4
```

```python
# Repeat the analysis and cross validation with the file bank-additional-full.csv.
# How does the performance of the model change (with the additional training
# examples and additional features)?

bankdatafull = pd.read_csv('bank-additional-full.csv',sep=';')
bankdatafull.head(10)
```

```python
bankdatafull.info()
```

```python
enc = OneHotEncoder()
categoricalvals = ['job','marital','education','default','housing','loan','contact','day_of_week','month','poutcome','y']
for col in categoricalvals:
    bankdatafull = text2int(bankdatafull, col)
    
bankdatafull.head()
```

```python
a = enc.fit_transform(bankdatafull[['job','marital','education','default','housing','loan','contact','day_of_week','month','poutcome']])  
dummies = pd.DataFrame(a.toarray()) 

X = pd.concat([bankdatafull[['age','duration','campaign','pdays','previous','emp.var.rate','cons.price.idx',
                             'cons.conf.idx','euribor3m','nr.employed']],dummies],axis=1)
y = bankdatafull.y.values

#3. Build a decision tree model to predict whether a prospect will buy the product.
# Using max depth of 3 here... 
from sklearn.tree import DecisionTreeClassifier
treeclf = DecisionTreeClassifier(max_depth=3, random_state=1) #random_state = seed value
treeclf.fit(X, y)

# compute the feature importance
pd.DataFrame({'feature':X.columns, 'importance':treeclf.feature_importances_}).sort('importance',ascending=False).reset_index().ix[:10]
```

```python
# Going to try out various max depths to see which one is best
from sklearn.cross_validation import cross_val_score
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

meanscores=[]
for md in range(1,6):
    treeclf = DecisionTreeClassifier(max_depth=md, random_state=3)
    scores = cross_val_score(treeclf, X, y, cv=5)
    meanscores.append(np.mean(scores))

plt.plot(range(1,6),meanscores)
plt.xlabel('Max Depth')
plt.ylabel('Scores')

# Including the additional features seems to actually reduce the model accuracy with a lower max depth leading to better results.
# We see that the feature importances are highest for nr.employed and cons.conf.idx above. Although it helps to separate the data,
# it appears to reduce the prediction power of the model. 
```
